# Git Basics for VS Code Users

This notebook is a beginner-friendly guide to the Git commands that cover what most people actually use in daily work.

The goal is simple:

- understand what Git is doing,
- learn the commands that solve about 90% of normal Git work,
- use Git comfortably inside VS Code and in the terminal.

This guide is written in very simple English.
You do not need to be an expert.
Follow slowly, step by step.


## 1. What Git is

Git is a tool that tracks changes in files.

Very simple idea:

- your project is a folder,
- Git watches the history of that folder,
- you can save snapshots,
- you can upload the project to GitHub,
- you can download updates,
- you can go back if something breaks.

Think like this:

- folder = your work
- Git = history machine
- GitHub = website that stores the Git project online


## 2. What VS Code does with Git

VS Code is not Git itself.

VS Code is just a nice interface that helps you use Git more easily.

In VS Code, you can:

- see changed files,
- stage files,
- write commit messages,
- commit,
- push,
- pull,
- resolve conflicts.

But under the hood, it is still Git.

That is why learning the terminal commands is still very useful.


## 3. The 90% Git workflow

These are the commands most people use again and again:

1. `git clone`
2. `git init`
3. `git status`
4. `git add`
5. `git commit`
6. `git push`
7. `git pull`
8. `git branch`
9. `git checkout` or `git switch`
10. `git merge`
11. `git log`
12. `git remote -v`

If you understand these well, you can already do most normal Git work.


## 4. Before you start

Check that Git is installed:

```bash
git --version
```

If you see a version number, Git is installed.

Set your name and email one time:

```bash
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```

Why this matters:

- Git needs your name and email for commits.
- Without this, `git commit` may fail.


## 5. Method A: Start a new Git project from a local folder

Use this when:

- you already have a folder on your computer,
- it is not yet a Git repository,
- you want to start tracking it with Git.

Step 1: move into the folder

```bash
cd /path/to/your/project
```

Step 2: start Git

```bash
git init
```

What this does:

- creates a hidden `.git` folder,
- tells Git to start tracking this project.

Step 3: check status

```bash
git status
```

This shows what Git currently sees.


## 6. Method B: Download an existing repository from GitHub

Use this when:

- the repository already exists on GitHub,
- you want a copy on your own computer.

Command:

```bash
git clone https://github.com/username/repository.git
```

Then go into the folder:

```bash
cd repository
```

What `git clone` does:

- downloads the full project,
- downloads the commit history,
- connects your local copy to the remote repository.


## 7. `git status` is your best friend

Command:

```bash
git status
```

Use this all the time.

Why:

- it shows changed files,
- it shows untracked files,
- it shows staged files,
- it shows which branch you are on.

If you feel lost, run `git status`.

This is the first command many experienced users run when confused.


## 8. Add files with `git add`

After changing files, Git sees them but does not save them yet.

You must add them first.

Add one file:

```bash
git add README.md
```

Add everything in the current folder:

```bash
git add .
```

Meaning:

- `git add` = prepare files for the next commit.

Important:

- `git add` does not upload anything,
- `git add` does not make a final save point,
- it only stages files.


## 9. Save work with `git commit`

After `git add`, make a commit.

Example:

```bash
git commit -m "Add README and update script"
```

Meaning:

- `git commit` = save a snapshot,
- `-m` = message,
- the text explains what changed.

Think of a commit like a clean save point in a game.

A good commit message should say what you changed.


## 10. Upload changes with `git push`

After committing, upload to GitHub:

```bash
git push
```

Sometimes for the first push you need:

```bash
git push -u origin main
```

Meaning:

- `git push` = send your local commits to the remote repository,
- `origin` = the default remote name,
- `main` = the branch.

Use `git push` after you commit changes and want GitHub updated.


## 11. Download updates with `git pull`

If the GitHub repository changed and you want the newest version on your computer:

```bash
git pull
```

Meaning:

- download remote changes,
- merge them into your local branch.

Use this before starting work if other people may have changed the project.


## 12. The normal daily cycle

This is the most common Git pattern:

```bash
git status
git add .
git commit -m "Describe your change"
git push
```

That is the daily workflow for many people.

If you remember only one pattern, remember this one.


## 13. See branches with `git branch`

Command:

```bash
git branch
```

This shows your branches.

The branch with `*` is your current branch.

Example:

```bash
* main
  dev
```

This means you are currently on `main`.


## 14. Change branch with `git switch`

Modern way:

```bash
git switch dev
```

Create and switch at the same time:

```bash
git switch -c new-feature
```

Older command you may still see:

```bash
git checkout dev
```

For beginners:

- prefer `git switch` for changing branches,
- `git checkout` is older and can do more things, which makes it a bit more confusing.


## 15. Why branches exist

A branch is like a parallel work line.

Use branches when:

- you want to test something safely,
- you want to build a feature without touching `main`,
- you want cleaner teamwork.

Common idea:

- `main` = stable version
- `feature-branch` = work in progress


## 16. Merge branch with `git merge`

After finishing work in a branch, merge it back.

Example:

```bash
git switch main
git merge new-feature
```

Meaning:

- go back to `main`,
- bring the changes from `new-feature` into `main`.


## 17. Check remote repository with `git remote -v`

Command:

```bash
git remote -v
```

This shows where your project pushes and pulls from.

Example:

```bash
origin  git@github.com:yourusername/project.git (fetch)
origin  git@github.com:yourusername/project.git (push)
```

Use this when:

- you are not sure which GitHub repo you are connected to,
- push goes to the wrong place,
- you copied a repository from somewhere else.


## 18. See commit history with `git log`

Command:

```bash
git log
```

Shorter version:

```bash
git log --oneline
```

This shows commit history.

Use `--oneline` if you want a cleaner, easier view.


## 19. Basic VS Code Git workflow

Inside VS Code, the usual workflow is:

1. open your project folder
2. click the Source Control tab
3. check changed files
4. stage files
5. write commit message
6. click Commit
7. click Sync Changes or Push

Important idea:

- VS Code buttons are just a friendlier face for Git commands.


## 20. How to open a Git project in VS Code

If the project already exists:

```bash
cd /path/to/project
code .
```

This opens the current folder in VS Code.

If VS Code does not recognize Git:

- check that `.git` exists in the folder,
- check that Git is installed,
- reload VS Code.


## 21. What the Source Control panel means in VS Code

In the Source Control tab:

- changed files = files Git sees as modified
- stage = prepare a file for commit
- commit message box = text that describes the save point
- Commit button = same idea as `git commit`
- Push/Sync = upload changes

Very important:

A file can be changed but not yet committed.
A file can be committed locally but not yet pushed.
These are different steps.


## 22. The most common beginner confusion

Many beginners think this:

- edit file in VS Code
- file is saved
- so GitHub should update automatically

That is wrong.

Saving a file is not the same as using Git.

The real Git flow is:

1. save file
2. `git add`
3. `git commit`
4. `git push`

All four ideas matter.


## 23. Example: upload a local project to GitHub

Imagine you already have a folder on your computer.

Step by step:

```bash
cd /path/to/project
git init
git add .
git commit -m "Initial commit"
git remote add origin git@github.com:YOUR_USERNAME/YOUR_REPO.git
git branch -M main
git push -u origin main
```

This is one of the most important real-life workflows.


## 24. Example: make a small change later

After the project already exists on GitHub, later you change one file.

Step by step:

```bash
git status
git add README.md
git commit -m "Update README"
git push
```

That is a very normal update cycle.


## 25. Example: create a branch for a new feature

```bash
git switch -c add-login-page
```

Now change files, then:

```bash
git add .
git commit -m "Create login page"
git push -u origin add-login-page
```

This uploads the new branch to GitHub.


## 26. Example: merge a finished branch

```bash
git switch main
git pull
git merge add-login-page
git push
```

Meaning:

- go back to main,
- get newest changes,
- merge the feature branch,
- upload result.


## 28. Common beginner errors

### Error: `author identity unknown`

Git does not know your name and email.

Fix:

```bash
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```

### Error: `remote origin already exists`

You already added a remote before.

Check:

```bash
git remote -v
```

Remove and add again if needed:

```bash
git remote remove origin
git remote add origin git@github.com:YOUR_USERNAME/YOUR_REPO.git
```

### Error: `src refspec main does not match any`

Usually means:

- you have not made a commit yet,
- or your branch name is different.

Fix:

```bash
git add .
git commit -m "Initial commit"
git branch -M main
git push -u origin main
```

### Error: `permission denied`

Usually means GitHub authentication is not set up correctly.


## 29. Good habits for beginners

1. Run `git status` often.
2. Write clear commit messages.
3. Pull before starting work on shared projects.
4. Push after important commits.
5. Use branches for risky changes.
6. Check `git remote -v` if something feels wrong.
7. Do not panic when Git shows many lines. Read slowly.


## 30. Recommended beginner command set to memorize

If you only memorize a small set, memorize these:

```bash
git status
git add .
git commit -m "message"
git push
git pull
git branch
git switch -c new-branch
git merge branch-name
git remote -v
git log --oneline
```

This covers a very large part of normal Git work.


## 31. Very simple mental model

Think of Git as 3 places:

1. working directory  
   your normal files

2. staging area  
   files prepared with `git add`

3. repository history  
   saved commits made with `git commit`

Then GitHub is the online copy, reached with `git push` and `git pull`.

This mental model makes many Git commands easier to understand.


## 32. Final beginner workflow for VS Code users

If you use VS Code and want the simplest practical workflow:

- open project in VS Code
- edit and save files
- look at Source Control
- stage files
- write commit message
- commit
- push

And when confused, use terminal:

```bash
git status
git branch
git remote -v
git log --oneline
```

These four commands answer many questions fast.
