# Guide: Push a Local Project from Your Computer to GitHub

This notebook explains, step by step, how to upload a project that already exists on your computer to a GitHub repository.

It is written for a complete beginner. Nothing is assumed. Follow slowly, one step at a time.


## 1. What you are trying to do

You already have a project folder on your computer.

For example:

```bash
~/Documents/my_project
```

You want to:

1. Turn that folder into a Git repository.
2. Create a new repository on GitHub.
3. Connect your local folder to that GitHub repository.
4. Push your files from your computer to GitHub.

Simple meaning:

- local project = the folder on your computer
- Git = tool that tracks versions of files
- GitHub = website that stores Git repositories online
- push = upload your local project to GitHub


## 2. Before you start

You need these things first:

- a GitHub account
- Git installed on your computer
- your project folder already exists on your machine

To check whether Git is installed, open Terminal and run:

```bash
git --version
```

If Git is installed, it will print a version number.


## 3. Example names used in this guide

We will use these examples:

- local project folder:

```bash
~/Documents/my_project
```

- GitHub username:

```bash
yourusername
```

- new GitHub repository name:

```bash
my_project
```

Replace these example names with your real names.


## 4. Step 1: Go into your project folder

Open Terminal.

Move into your project folder:

```bash
cd ~/Documents/my_project
```

Now you are standing inside the folder that you want to upload.

To make sure you are in the right place, run:

```bash
pwd
ls
```

`pwd` shows where you are.
`ls` shows the files in the folder.


## 5. Step 2: Turn the folder into a Git repository

Run:

```bash
git init
```

What this does:

- creates a hidden `.git` folder inside your project
- tells Git: "start tracking this folder"

After this, your folder becomes a Git repository.


## 6. Step 3: Check Git status

Run:

```bash
git status
```

This tells you what Git sees.

At the beginning, Git will usually say that files are untracked.
That is normal.
It just means Git sees the files, but has not started saving them yet.


## 7. Step 4: Add files to Git tracking

Run:

```bash
git add .
```

Meaning:

- `git add` = tell Git which files should be included
- `.` = include everything in the current folder

This does not upload anything yet.
It only prepares the files for a commit.


## 8. Step 5: Save your first snapshot with a commit

Run:

```bash
git commit -m "Initial commit"
```

Meaning:

- `git commit` = save a snapshot of the current project state
- `-m` = message
- `"Initial commit"` = the commit message

Think of a commit as a save point.

If Git complains that your name or email is not set, use these commands first:

```bash
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```

Then run the commit command again.


## 9. Step 6: Create a new repository on GitHub

Go to GitHub website and do this:

1. Sign in.
2. Click **New repository**.
3. Enter repository name, for example `my_project`.
4. Choose **Public** or **Private**.
5. Do not add README, `.gitignore`, or license if your local project already has files and you want the cleanest push.
6. Click **Create repository**.

After creating it, GitHub will show a repository URL.

SSH version may look like:

```bash
git@github.com:yourusername/my_project.git
```

HTTPS version may look like:

```bash
https://github.com/yourusername/my_project.git
```


## 10. Step 7: Connect your local Git project to GitHub

Still inside your project folder, run one of these commands.

### If using SSH

```bash
git remote add origin git@github.com:yourusername/my_project.git
```

### If using HTTPS

```bash
git remote add origin https://github.com/yourusername/my_project.git
```

Meaning:

- `git remote add` = save an online repository address
- `origin` = the standard name for your main remote repository
- the GitHub URL = where your local project will be uploaded


## 11. Step 8: Check that the remote was added correctly

Run:

```bash
git remote -v
```

You should see something like:

```bash
origin  git@github.com:yourusername/my_project.git (fetch)
origin  git@github.com:yourusername/my_project.git (push)
```

or the HTTPS version.

This means your local project now knows which GitHub repository it belongs to.


## 12. Step 9: Choose or rename the main branch

Many repositories use `main` as the default branch.

To make your branch name `main`, run:

```bash
git branch -M main
```

Meaning:

- `git branch` = work with branches
- `-M` = rename branch forcefully if needed
- `main` = the new branch name


## 13. Step 10: Push your local project to GitHub

Run:

```bash
git push -u origin main
```

Meaning:

- `git push` = upload local commits to GitHub
- `-u` = remember this remote branch for future pushes
- `origin` = your GitHub repository
- `main` = the branch name

After this finishes, refresh your GitHub page.
Your files should now be visible there.


## 14. The shortest working command sequence

If your project folder already exists and you want the minimal useful version, here it is:

```bash
cd ~/Documents/my_project
git init
git add .
git commit -m "Initial commit"
git remote add origin git@github.com:yourusername/my_project.git
git branch -M main
git push -u origin main
```

Replace the path, username, and repository name with your real ones.


## 15. What to do next time after you change files

After your project is already connected to GitHub, future updates are simpler.

Use:

```bash
git add .
git commit -m "describe what changed"
git push
```

Because of `-u origin main` from before, later you usually only need `git push`.


## 16. Common errors and how to understand them

### Error: author identity unknown

Git does not know your name and email yet.
Fix it with:

```bash
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```

### Error: remote origin already exists

You already added a remote before.
Check it:

```bash
git remote -v
```

If you need to replace it:

```bash
git remote remove origin
git remote add origin git@github.com:yourusername/my_project.git
```

### Error: src refspec main does not match any

Usually means no commit exists yet, or branch name is wrong.

Fix:

1. make sure you already ran `git commit`
2. then run:

```bash
git branch -M main
git push -u origin main
```

### Error: permission denied

Usually means your GitHub authentication is not set up correctly.

If using SSH, check your SSH key setup.
If using HTTPS, GitHub may ask for sign-in or a personal access token.

### Error: repository not found

Usually means:

- the repository URL is wrong,
- the repository was not created yet,
- or you do not have access.


## 17. How to check everything if you feel lost

These commands help you inspect the current situation:

```bash
pwd
ls
git status
git branch
git remote -v
```

What they tell you:

- `pwd` = where you are
- `ls` = what files are here
- `git status` = what Git thinks about your folder
- `git branch` = current branch name
- `git remote -v` = which GitHub repo your project points to


## 18. Recommended safe beginner workflow

Use this version if you want the cleanest beginner path:

```bash
cd ~/Documents/my_project
git init
git add .
git commit -m "Initial commit"
git remote add origin git@github.com:yourusername/my_project.git
git branch -M main
git push -u origin main
```

This is enough for most normal cases.


## 19. Very short summary

The basic idea is:

1. go into your local folder
2. start Git there
3. add files
4. commit files
5. create GitHub repo
6. connect local repo to GitHub repo
7. push

That is the whole workflow.
