# README Guide: Clone a Public GitHub Repository and Push It to Your Own Private Repository

This guide explains, step by step, how to:

1. Clone another person's GitHub repository to your computer.
2. Create a new private repository on your own GitHub account.
3. Push the cloned project into your own private repository.

This is written for beginners. Every command is explained. Follow it slowly, one step at a time.


## 1. What you are trying to do

You have a public repository from someone else, for example:

```bash
https://github.com/otheruser/projectA.git
```

You want to:

- download it to your own computer,
- then upload that content into a new repository on your own GitHub account,
- and keep your new repository private.

Important:

- `git clone` only downloads the repository to your computer.
- It does not create a repository on your GitHub account.
- To make it appear in your own GitHub account, you must create your own repository and push to it.


## 2. Before you start

Make sure you already have:

- a GitHub account,
- Git installed on your computer,
- permission to use the source repository according to its license,
- either SSH or HTTPS set up for pushing to GitHub.

To check whether Git is installed, run:

```bash
git --version
```

If Git is installed, you will see a version number.


## 3. Example used in this guide

In this guide, we will pretend:

- the source repository is:

```bash
https://github.com/otheruser/projectA.git
```

- your GitHub username is:

```bash
yourusername
```

- your new private repository name will be:

```bash
my-private-project
```

Replace these example names with your real ones.


## 4. Step 1: Clone the other person's repository to your computer

Run:

```bash
git clone https://github.com/otheruser/projectA.git
```

What this does:

- downloads the repository to your computer,
- creates a folder named after the repository,
- keeps the full Git history.

Then move into that folder:

```bash
cd projectA
```

Now you are inside the cloned repository on your own machine.


## 5. Step 2: Check the current remote

Run:

```bash
git remote -v
```

You may see something like this:

```bash
origin  https://github.com/otheruser/projectA.git (fetch)
origin  https://github.com/otheruser/projectA.git (push)
```

Meaning:

- `origin` is the saved address of the current remote repository.
- Right now, your local copy is linked to the other person's repository.

You usually cannot push there, because it is not your repository.


## 6. Step 3: Create your own new private repository on GitHub

On GitHub website:

1. Sign in to your GitHub account.
2. Click **New repository**.
3. Repository name: `my-private-project`
4. Choose **Private**.
5. Do not add README, `.gitignore`, or license if you want to push the cloned project cleanly.
6. Click **Create repository**.

After creating it, GitHub will show you the repository URL.

It may look like one of these:

SSH:

```bash
git@github.com:yourusername/my-private-project.git
```

HTTPS:

```bash
https://github.com/yourusername/my-private-project.git
```


## 7. Step 4: Connect your local copy to your own private repository

There are two common ways to do this.

### Option A: Replace the old remote completely

Use this if you do not care about keeping the old repository link saved.

```bash
git remote remove origin
git remote add origin git@github.com:yourusername/my-private-project.git
```

If you use HTTPS instead of SSH:

```bash
git remote remove origin
git remote add origin https://github.com/yourusername/my-private-project.git
```

### Option B: Keep the original repository link as `upstream`

This is cleaner and more professional. It lets you still fetch updates from the original repository later.

```bash
git remote rename origin upstream
git remote add origin git@github.com:yourusername/my-private-project.git
```

If using HTTPS:

```bash
git remote rename origin upstream
git remote add origin https://github.com/yourusername/my-private-project.git
```

Meaning:

- `upstream` = the original source repository,
- `origin` = your own private repository.

This is usually the best choice.


## 8. Step 5: Check your remotes again

Run:

```bash
git remote -v
```

If you used Option B, you should see something like:

```bash
origin    git@github.com:yourusername/my-private-project.git (fetch)
origin    git@github.com:yourusername/my-private-project.git (push)
upstream  https://github.com/otheruser/projectA.git (fetch)
upstream  https://github.com/otheruser/projectA.git (push)
```

That means your local repository is now set up correctly.


## 9. Step 6: Find the current branch name

Run:

```bash
git branch
```

You may see:

```bash
* main
```

or:

```bash
* master
```

The line with `*` is your current branch.

You must push that branch.


## 10. Step 7: Push the repository to your own private GitHub repository

If your branch is `main`, run:

```bash
git push -u origin main
```

If your branch is `master`, run:

```bash
git push -u origin master
```

Meaning of this command:

- `git push` = upload your local commits to the remote repository,
- `-u` = set this remote branch as the default upstream for later pushes,
- `origin` = your private repository,
- `main` or `master` = the branch name.

After this, refresh your GitHub page.
You should now see the full project inside your own private repository.


## 11. Full command sequence, minimal version

If you want the shortest useful workflow, here it is.

### SSH version

```bash
git clone https://github.com/otheruser/projectA.git
cd projectA
git remote rename origin upstream
git remote add origin git@github.com:yourusername/my-private-project.git
git branch
git push -u origin main
```

If the branch is not `main`, replace `main` with your real branch name.


## 12. Full command sequence, HTTPS version

```bash
git clone https://github.com/otheruser/projectA.git
cd projectA
git remote rename origin upstream
git remote add origin https://github.com/yourusername/my-private-project.git
git branch
git push -u origin main
```

If GitHub asks for authentication, follow the login flow or use a personal access token if required.


## 13. If you want a completely fresh repository with no old commit history

Sometimes you want only the files, not the original commit history.

Then do this instead:

```bash
git clone https://github.com/otheruser/projectA.git
cd projectA
rm -rf .git
git init
git add .
git commit -m "Initial commit"
git branch -M main
git remote add origin git@github.com:yourusername/my-private-project.git
git push -u origin main
```

What this does:

- removes the old Git history,
- turns the project into a brand new repository,
- makes your repository start from your own first commit.


## 14. If you want to keep getting updates from the original repository later

This is why Option B is useful.

If you kept the original repo as `upstream`, later you can run:

```bash
git fetch upstream
```

Then merge from the original repository:

```bash
git merge upstream/main
```

If the original repository uses `master` instead of `main`, then use:

```bash
git merge upstream/master
```

This lets you keep your private version while still pulling new changes from the source repository.


## 15. Common errors and what they mean

### Error: permission denied when pushing

Cause:

- you are still pointing to the other person's repository,
- or your GitHub authentication is not set up.

Check:

```bash
git remote -v
```

Make sure `origin` points to your own private repository.

### Error: repository not found

Cause:

- the repository URL is wrong,
- or the private repository has not been created yet,
- or your account does not have access.

### Error: src refspec main does not match any

Cause:

- your current branch is not named `main`.

Run:

```bash
git branch
```

Then push the correct branch name.

### Error: remote origin already exists

Cause:

- `origin` is already defined.

Fix:

```bash
git remote remove origin
```

or rename it:

```bash
git remote rename origin upstream
```


## 16. A safe and professional recommended workflow

For most people, this is the best version to use:

```bash
git clone https://github.com/otheruser/projectA.git
cd projectA
git remote rename origin upstream
git remote add origin git@github.com:yourusername/my-private-project.git
git remote -v
git branch
git push -u origin main
```

Why this is good:

- you keep the original source saved as `upstream`,
- your own repository is clearly marked as `origin`,
- you can still pull updates from the original repository later,
- your own repository stays private.


## 17. Very short summary

Clone means:

- copy repository from GitHub to your computer.

Push means:

- upload your local repository to a remote GitHub repository.

Private repository means:

- only you and people you allow can see it.

Best beginner idea:

1. Clone the source repository.
2. Create your own private repository.
3. Point `origin` to your own repository.
4. Push.
